```python
import seaborn as sns
import pandas as pd
```

# Language Modeling

```python
results = [
    ['2-gram','Normal Text',515.89],
    ['2-gram','Consonants',451.74],
    ['2-gram','Masked Consonants',481.18],
    ['3-gram','Normal Text',444.02],
    ['3-gram','Consonants',388.08],
    ['3-gram','Masked Consonants',411.84],
    ['4-gram','Normal Text',434.92],
    ['4-gram','Consonants',379.16],
    ['4-gram','Masked Consonants',402.76],
    ['5-gram','Normal Text',433.22],
    ['5-gram','Consonants',377.49],
    ['5-gram','Masked Consonants',401.01],
    ['6-gram','Normal Text',432.90],
    ['6-gram','Consonants',377.21],
    ['6-gram','Masked Consonants',400.73],
    ['Nerual','Normal Text',102.47],
    ['Nerual','Consonants',106.83],
    ['Nerual','Masked Consonants',100.29],
]
```

```python
df = pd.DataFrame(results)
df.columns = ['Model', 'Text Type', 'Perplexity']
df
```

```python
sns.set(rc={'figure.figsize':(12,7)})
sns.set_style("whitegrid")
g = sns.barplot(
    x="Text Type",
    y="Perplexity",
    hue='Model',
    data=df,
  )
sns.move_legend(g,loc='upper right')
```

# Text Classification
